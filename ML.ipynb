```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_circles
```

```python
#формирую необходимую выборку и вывожу её график 
x1, y1 = make_moons(n_samples=200, noise=0.01, random_state=0)
plt.scatter(x1[:, 0], x1[:, 1], c=y1)
plt.show()
```

```python
#формируем обучающий и тестовый наборы данных
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1,y1, test_size=0.2, train_size=0.8, random_state = 0)
print(f'x1_train shape: {x1_train.shape}')
print(f'x1_test shape: {x1_test.shape}')
print(f'y1_train shape: {y1_train.shape}')
print(f'y1_test shape: {y1_test.shape}')
```

```python
#создаю график точности модели в зависимости от количества классов 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

k_values = range(1, 9)
accuracy_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x1_train, y1_train)
    y1_pred = knn.predict(x1_test)
    accuracy = accuracy_score(y1_test, y1_pred)
    accuracy_scores.append(accuracy)

plt.plot(k_values, accuracy_scores)
plt.xlabel('k')
plt.ylabel('Точность')
plt.title('Точность модели классификации KNN')
plt.show()
#количество классов не влияет на точность 
```

```python
#рассчитываю необходимые метрики 
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracy = accuracy_score(y1_test, y1_pred)
precision = precision_score(y1_test, y1_pred)
recall = recall_score(y1_test, y1_pred)
f1 = f1_score(y1_test, y1_pred)
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1: {f1}')
```

```python
#вывожу матрицу ошибок
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(y1_test, y1_pred)
print(matrix)
```

```python
x = np.random.uniform(-10, 10, 150)
y = np.random.uniform(-10, 10, 150)
class1 = (x < 0) & (y < 0) | (x > 0) & (y > 0)
class2 = (x < 0) & (y > 0) | (x > 0) & (y < 0)
dataset = np.array([x, y, class1, class2]).T
plt.scatter(x, y, c=class1, s=50, alpha=0.3)
plt.scatter(x, y, c=class2, s=50, alpha=0.3)
plt.show()
```

```python
train_set, test_set = train_test_split(dataset, test_size=0.2, train_size=0.8, random_state = 0)
print(f'train_set shape: {train_set.shape}')
print(f'test_set: {test_set.shape}')
```

```python
k_values = range(1, 9)
accuracy_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_set[:, :2], train_set[:, 2])
    y4_pred = knn.predict(test_set[:, :2])
    accuracy = accuracy_score(test_set[:, 2], y4_pred)
    accuracy_scores.append(accuracy)

plt.plot(k_values, accuracy_scores)
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Classification Accuracy of KNN Model')
plt.show()
```

```python
accuracy = accuracy_score(test_set[:, 2], y4_pred)
precision = precision_score(test_set[:, 2], y4_pred)
recall = recall_score(test_set[:, 2], y4_pred)
F1 = f1_score(test_set[:, 2], y4_pred)
print(f'Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1: {F1}')
```
